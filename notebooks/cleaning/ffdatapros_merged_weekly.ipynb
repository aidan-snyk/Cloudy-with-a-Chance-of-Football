```python
import pandas as pd
import missingno as msno
df = pd.read_csv ('C:/Users/Yaphet Tewahade/Downloads/fantasyfootballpros_raw_week1.csv')
```

```python
shape_info = df.shape
print('This dataset evaluates the performance of {} nfl players based on their week 1 actual fantasy points from {} columns' .format(shape_info[0], # using the .format method, insert the number of rows in the dataframe
              shape_info[1]))
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
list(df)
```

```python
df_old_cols = list(df)

df_new_cols = [x.replace(" ", "").capitalize() for x in df_old_cols]
```

```python
df.columns = df_new_cols
df.head(2)
```

```python
msno.matrix(df)
```

```python
df = df.fillna(0)
```

```python
df.isnull().values.any() #checks for any null values
```

```python
df.to_csv('C:/Users/Yaphet Tewahade//Downloads/fantasyfootballpros_2019_week1_cleaned.csv')
```
